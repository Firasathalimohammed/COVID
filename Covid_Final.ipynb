The `CovidDataScraper` class is a specialized Python tool designed for extracting real-time COVID-19 data from the Worldometer website. This class utilizes libraries such as `requests` for web requests and `BeautifulSoup` from `bs4` for parsing HTML data, making it an effective tool for web scraping.

### Key Components of the CovidDataScraper Class:

1. **Initialization (`__init__` method)**:
   - The class initializes with a predefined URL (`https://www.worldometers.info/coronavirus/`), which is the source of the COVID-19 data.
   - It also defines a `field_list` attribute, which contains the names of the data fields to be scraped, such as "Total Cases", "New Cases", "Total Deaths", and so on.

2. **Data Fetching (`fetch_data` method)**:
   - This method sends a GET request to the Worldometer URL and parses the received HTML text using `BeautifulSoup`.
   - It then extracts the data from specific parts of the HTML (table rows within the body) and stores the data in a list.
   - The method processes each row of the table, extracting and stripping text from the relevant columns (defined by `field_list`).
   - Finally, it returns the data as a Pandas DataFrame, making it suitable for further data analysis tasks.

3. **Global Statistics Retrieval (`get_global_stats` method)**:
   - Similar to `fetch_data`, this method fetches and parses the HTML from the Worldometer page.
   - It specifically targets the "maincounter-number" class divs, which contain global statistics like "Total Cases", "Total Deaths", and "Total Recovered".
   - The method extracts and strips text from these elements, returning a dictionary of the global COVID-19 statistics.

### Usage:
- An instance of the `CovidDataScraper` class is created.
- The `fetch_data` method is called to scrape country-specific COVID-19 data, which is then stored in a Pandas DataFrame.
- The `get_global_stats` method is invoked to retrieve worldwide COVID-19 statistics.
- The country-specific data can be exported to a CSV file for further use or analysis.

This class serves as a crucial component in a larger data analysis pipeline, especially for tasks requiring up-to-date information on the COVID-19 pandemic. The ability to programmatically extract and process data from a web source makes it a valuable tool for researchers and analysts working in public health, policy planning, and related fields.

# **Research Question: "How do COVID-19 case and death rates correlate with various demographic and health indicators across different countries?"**

This question aims to understand the relationship between COVID-19 impacts (measured in terms of case and death rates) and various country-specific indicators such as median age, GDP per capita, hospital beds per thousand people, diabetes prevalence, and population density. The analysis will involve statistical techniques to explore correlations and visualizations to present these relationships

```python
# Calculate case and death rates per million
cleaned_df['case_rate_per_million'] = cleaned_df['total_cases'] / cleaned_df['population'] * 1e6
cleaned_df['death_rate_per_million'] = cleaned_df['total_deaths'] / cleaned_df['population'] * 1e6

# Select relevant columns for correlation analysis
columns_of_interest = ['case_rate_per_million', 'death_rate_per_million', 'median_age',
                       'gdp_per_capita', 'hospital_beds_per_thousand', 'diabetes_prevalence',
                       'population_density']
correlation_data = cleaned_df[columns_of_interest].corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of COVID-19 Rates and Demographic/Health Indicators')
plt.show()

# Scatter plots to visualize the relationships
# This is an example for case rate and GDP per capita
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cleaned_df, x='gdp_per_capita', y='case_rate_per_million', hue='continent', alpha=0.6)
plt.title('COVID-19 Case Rate vs GDP per Capita by Continent')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('COVID-19 Case Rate per Million')
plt.show()
```

1. **Correlation Matrix of COVID-19 Rates and Demographic/Health Indicators**:
   - The heatmap visualizes the correlation coefficients between different variables such as case rate, death rate, median age, GDP per capita, hospital beds per thousand, diabetes prevalence, and population density.
   - Positive values (closer to 1.0) indicate a positive correlation, meaning as one variable increases, the other tends to increase as well. Negative values (closer to -1.0) indicate a negative correlation, where one variable increases as the other decreases.
   - A strong correlation between case rate and death rate would not be surprising, as regions with high numbers of COVID-19 cases often experience higher mortality rates.
   - The correlations between demographic factors like median age or GDP per capita and COVID-19 rates could reveal socioeconomic factors influencing the pandemic's impact.
   - An absence of correlation (values around 0) suggests no linear relationship between the variables.

2. **COVID-19 Case Rate vs GDP per Capita by Continent**:
   - The scatter plot compares the case rate per million against GDP per capita for different continents.
   - Each point represents a specific country, with its position on the graph indicating both its GDP per capita and its COVID-19 case rate.
   - A pattern where countries with higher GDP per capita have lower or higher case rates could suggest the influence of economic factors on the spread or control of the virus.
   - The color coding by continent allows for regional comparisons, possibly highlighting how different parts of the world are affected by the pandemic in the context of their economic status.



# **Research Question: "How has the weekly trend of COVID-19 new cases and deaths evolved over time for select countries, and how do these trends compare?"**

This question will focus on analyzing the temporal evolution of COVID-19 in terms of new cases and deaths on a weekly basis. To answer this, we'll reshape the dataset from wide to long format, focusing on weekly data for a clearer temporal analysis

```python
# Reshaping data from wide to long format
long_df = cleaned_df.melt(id_vars=['iso_code', 'continent', 'location', 'date'],
                          value_vars=['new_cases', 'new_deaths'],
                          var_name='type', value_name='count')

# Filtering for specific countries, e.g., USA, India, Brazil
countries_of_interest = ['USA', 'IND', 'BRA']
filtered_df = long_df[long_df['location'].isin(countries_of_interest)]

# Creating a weekly data frame
filtered_df['week'] = filtered_df['date'].dt.isocalendar().week
weekly_df = filtered_df.groupby(['location', 'type', 'week']).sum().reset_index()

# Plotting the trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=weekly_df, x='week', y='count', hue='location', style='type', markers=True)
plt.title('Weekly COVID-19 Cases and Deaths for Selected Countries')
plt.xlabel('Week Number')
plt.ylabel('Count')
plt.legend(title='Country & Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
# Convert 'date' to datetime and create new features
cleaned_df['date'] = pd.to_datetime(cleaned_df['date'])
cleaned_df['new_cases_change'] = cleaned_df.groupby('location')['new_cases'].pct_change()
cleaned_df['new_deaths_change'] = cleaned_df.groupby('location')['new_deaths'].pct_change()

# Replace infinities with NaN and drop rows with NaN in target columns
cleaned_df.replace([np.inf, -np.inf], np.nan, inplace=True)
cleaned_df.dropna(subset=['new_cases', 'new_deaths'], inplace=True)

# Define numerical and categorical features
numerical_features = ['new_cases_change', 'new_deaths_change', 'population_density', 'gdp_per_capita', 'total_vaccinations_per_hundred']
categorical_features = ['continent', 'location']

# Preprocessing pipelines for both numerical and categorical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Split data into features and target variable
X = cleaned_df.drop(['new_cases', 'new_deaths'], axis=1)
y = cleaned_df[['new_cases', 'new_deaths']]

# Apply the preprocessor and split the dataset
X_prepared = preprocessor.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size=0.3, random_state=42)

# Model Training
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Making predictions
predictions = model.predict(X_test)

# Display the first few predictions
print(predictions[:5])
```
